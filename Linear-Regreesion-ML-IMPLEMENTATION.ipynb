# Standardizing a dataset
It involves rescaling the distribution of values so that means of obsereved values
is o and the standard deviation is 1 .

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
X_train
```

# why we standardised a dataset in linear regreesion

A variable is standardized by subtracting from it its sample mean and by dividing it by its standard deviation. After being standardized, the variable has zero mean and unit standard deviation.
linear regressions = In regression analysis, you need to standardize the independent variables when your model contains polynomial terms to model curvature or interaction terms. These terms provide crucial information about the relationships between the independent variables and the dependent variable, but they also generate high amounts of multicollinearity.

# Model Trainning

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
## print the coefficients and the intercepts
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
## on which parameters the model has been trained 
regression.get_params()
```

# Prediction with Test Data

```python
reg_pred=regression.predict(X_test)
```

```python
reg_pred
```

# To be sure about our Model  we need to some performance matrix

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(y_test,reg_pred))
print(mean_squared_error(y_test,reg_pred))
print(np.sqrt(mean_squared_error(y_test,reg_pred)))
```

Mean Absolute Error (MAE) is the sum of the absolute difference between actual and predicted values
Mean Squared Error is calculated by taking the average of the square of the difference between the original and predicted values of the data.

# R square and adjusted R square 

R Squared is used to determine the strength of correlation between the predictors and the target. In simple terms it lets us know how good a regression model is when compared to the average. R Squared is the ratio between the residual sum of squares and the total sum of squares.

R Squared has no relation to express the effect of a bad or least significant independent variable on the regression. Thus even if the model consists of a less significant variable say, for example, the personâ€™s Name for predicting the Salary, the value of R squared will increase suggesting that the model is better.

Compared to R Squared which can only increase, Adjusted R Squared has the capability to decrease with the addition of less significant variables, thus resulting in a more reliable and accurate evaluation.

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,reg_pred)
print(score)
```
